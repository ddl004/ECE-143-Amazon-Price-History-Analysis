```python
import keepa
import numpy as np
import matplotlib as plt

api_key = 'e6ihvarndmd2iee2bgeg60afm06gru9242g310tb4tv1kji72u57uon4us908d5h'

api = keepa.Keepa(api_key)
```

```python
bestsellers = []
software = '229534'
bestsellers = api.best_sellers_query(software)

print(bestsellers[0])
```

```python
bestsellers
```

```python
api.category_lookup('229534')
```

```python
print(len(bestsellers))
np.save('bestsellers.npy', bestsellers)
```

```python
print(bestsellers[0:50])
```

```python
# Save Product Dictionaries to a file

products_software_bs = api.query(bestsellers[0:200])
assert(len(products_software_bs) > 50)
assert(type(products_software_bs[0]) == dict)
np.save('product_software.npy', products_software_bs)
```

```python
# To load a product dictionary:
products_software_bs = np.load('product_software.npy', allow_pickle=True)
```

```python
# Count how many products have a price history longer than 50

cnt = 0
for product in products_software_bs:
    if len(product['data']['AMAZON']) > 50:
        cnt += 1
print(cnt)
```

```python
# Example plot
keepa.plot_product(products_software_bs[25])
```
