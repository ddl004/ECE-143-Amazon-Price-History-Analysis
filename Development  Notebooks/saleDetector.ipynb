```python
import numpy as np 
```

```python
products_toys_d = np.load('product_toys.npy', allow_pickle=True)
```

```python
def saleDetector(product):
    """
    To detect the sale price and timing, and sale 
    is definded as lower than avgerage*0.7
    (The architecture of the same function in product class)
    :param:product
    :type: dict
    :return: two list that are price and time 
    """
    data = product['data']
    times = data['AMAZON_time']
    price = data['AMAZON']
    assert len(times) == len(price)
    
    meanValue = np.nanmean(price);
    sale_price = list()
    sale_time = list()
    for i in range(len(price)):
        if price[i] < meanValue*0.7:
            sale_price.append(price[i])
            sale_time.append(times[i])
    return sale_price, sale_time
    
    
```

```python
a = saleDetector(products_toys_d[25])
```
